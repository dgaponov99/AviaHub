```python
import json
import requests
```

```python
map_file = "map.png"
```

```python
link = 'https://static-maps.yandex.ru/1.x/?l=map'
z = 'z=9'
bbox = 'bbox=75.40%2C53.32~84.514253,57.142546'
```

```python
border = '&pl=c:cf59b1FF,f:e6a8d71F,w:2,' \
            '75.689103%2C54.122893,75.118851%2C55.714225,76.303929%2C56.172764,76.094848%2C57.235334,' \
            '79.167280%2C57.004970,82.029356%2C56.266215,83.130449%2C56.538112,83.424866%2C55.677703,' \
            '84.441649%2C56.042187,85.115917%2C54.437695,84.444094%2C54.141674,84.208184%2C54.189280,' \
            '83.202192%2C54.061655,82.500403%2C53.496866,80.914809%2C54.320967,77.907677%2C53.290318,' \
            '76.502777%2C54.043163,76.915115%2C54.455445,75.689103%2C54.122893'
```

```python
with open("coordinates_of_district_center.json", "r") as read_file:
    data = json.load(read_file)
```

```python
pt = 'pt='
data = data.values()
for coord in data:
    pt += coord[0] + ',' + coord[1] + ',pm2gnm~'
pt = pt[:-1]
```

```python
map = requests.get(link + '&' + z + '&' + bbox + '&' + pt + '&' + border)
```

```python
with open(map_file, "wb") as file:
    file.write(map.content)
```
