```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import json
```

```python
plt.figure(figsize=(12, 9))
plt.axis([0, 600, 450, 0])
image = mpimg.imread("map.png")
plt.imshow(image)
plt.show()
```

```python
def coor2plt(coor):
    return (float(coor[0]) - 73.4) * (600/13.16), 450 - (float(coor[1]) - 52.4)*(450/5.63)
```

```python
with open("coordinates_of_district_center.json", "r") as read_file:
    data = json.load(read_file)
```

```python
plt.figure(figsize=(12, 9))
plt.axis([0, 600, 450, 0])
image = mpimg.imread("map.png")
plt.imshow(image)

new_data = {}

keys = data.keys()
for key in keys:
    x, y = coor2plt(data[key])
    new_data[key] = [x, y, data[key][-1]]
    plt.plot(x, y, '*')
    
plt.show()
```

```python
with open("new_coordinates_of_district_center.json", "w") as write_file:
    json.dump(new_data, write_file)
```
